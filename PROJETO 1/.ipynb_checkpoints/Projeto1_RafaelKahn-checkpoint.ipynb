# Introdução:

Desde sua fundação no dia 30 de Setembro de 1961, a Organização para a Cooperação e Desenvolvimento Econômico (OCDE) tem como objetivo formentar e incentivar o desenvolvimento econômico de seus países-membros, além de propor medidas que visam o equilíbrio econômico mundial. Esta organização internacional é composta atualmente por 36 membros (Como por exemplo Estados Unidos, Canadá, Reino Unido, entre outros), tendo seus países-membros considerados como países de renda alta pelo Banco Mundial.
Muitos membros da OCDE estão presentes na organização desde a sua fundação, porém existem membros que só começaram a fazer parte da mesma mais recentemente, tendo países como candidatos para ser o 37º país da organização como é o caso da Colombia.
Atualmente, a OCDE e seus parceiros-chave representam 80% dos invesimentos e transações mundiais, fato este que demonstra a importância econômica desta organização.

# Pergunta abordada:
Países que estão na OCDE tem um crescimento do PIB anual melhor que os países pertencentes à MERCOSUL?


# Hipótese:
Participar da OCDE é mais vantajoso que participar do MERCOSUL, pois por possuir membros com maior poder econômico que os membros da MERCOSUL e também pelo fato de que a OCDE é uma organização com um número de membros bem maior que o MERCOSUL, ela acaba gerando melhores ofertas econômicas para seus membros e fornece mais opções que as que organização sul-americana consegue fornecer. Com isso, o efeito que a OCDE tem sobre o PIB de seus membros também acaba se tornando maior, o que resulta assim em um melhor crescimento do PIB de seus membros quando comparado ao crescimento do PIB dos países do MERCOSUL.

# Análise:

```python
#Bibliotecas
%matplotlib inline

import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
#Leitura dos arquivos
dados = pd.read_excel("gdp_total_yearly_growth.xlsx")
```

```python
#Criando grupo OCDE
ocde = ('Australia', 'Austria', 'Belgium', 'Canada', 'Chile', 'Czech Republic', 'Denmark', 'Estonia', 'Finland','France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Israel', 'Italy', 'Japan', 'Latvia', 'Lithuania', 'Luxembourg', 'Mexico', 'Netherlands', 'New Zeland', 'Norway', 'Poland', 'Portugal', 'Slovak Republic', 'Slovenia', 'South Korea', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom', 'United States')
mercosul = ('Argentina', 'Brazil', 'Uruguay', 'Paraguay')
dados['Grupos']=0
for p in ocde:
    dados.loc[dados.country==p, 'Grupos']=1
for i in mercosul:
    dados.loc[dados.country==i, 'Grupos']=2
```

##### Legenda:
0 = Resto do Mundo<br>
1 = OCDE<br>
2 = MERCOSUL

Para que seja possível responder a pergunta abordada durante este projeto, é necessário com que analisemos dados do crescimento econômico do membros da OCDE e dos países pertencentes à MERCOSUL, que são eles: Brasil, Argentina, Uruguai e Paraguai. Fazendo isso, conseguimos obter boas comparações no desenvolvimento do PIB de cada grupo e o quanto fazer parte da OCDE influencia na economia de um país comparado a influência que o MERCOSUL exerce em seus países-membros.<br>
Os dados abordados durante este projeto serão do período de 1961 até 2013.

### Países membros da OCDE:

```python
dados_ocde = dados[(dados.Grupos==1)]
dados_ocde
```

## Países membros do MERCOSUL:

```python
dados_mercosul = dados[(dados.Grupos==2)]
dados_mercosul
```

Vamos utilizar apenas os dados do período entre 1961 e 2013, então é necessário filtrar novamente nossa base de dados para visualizarmos apenas os dados desejados.

```python
dados_ocde.iloc[:, np.r_[0, 161:215]]
```

```python
dados_mercosul.iloc[:, np.r_[0, 161:215]]
```

Para visualizarmos melhor o crescimento dos países de cada grupo, é necessário fazer uma média o crescimento anual do PIB entre os membros destes grupos, para assim obtermos dados e valores que podem contribuir em nossa análise tornando possível a comprovação de nossa hipótese. 

```python
dados_media = dados.groupby('Grupos').mean()
dados_media_filtrado = dados_media.iloc[:, np.r_[160:213]]
dados_media_filtrado
```

##### Legenda:
0 = Resto do Mundo<br>
1 = OCDE<br>
2 = MERCOSUL

Após conseguir os valores médios do crescimento anual das duas organizações, é possível visualizar a diferença entre as duas organizaões através de gráficos:

```python
#Gráfico Resto do Mundo
tempo = np.arange(1961, 2014)
plot_0=dados_media_filtrado.iloc[0, :]
plt.plot(tempo, plot_0, 'red', label='Resto do Mundo')
plt.xlabel('Tempo(Anos)')
plt.ylabel('Crescimento do PIB(%)')
plt.grid(True)
plt.legend()
plt.title('Crescimento anual do PIB x Tempo')
plt.show()
```

```python
#Gráfico OCDE
plot_1=dados_media_filtrado.iloc[1, :]
plt.plot(tempo, plot_1, 'blue', label='OCDE')
plt.xlabel('Tempo(Anos)')
plt.ylabel('Crescimento do PIB(%)')
plt.grid(True)
plt.legend()
plt.title('Crescimento anual do PIB x Tempo')
plt.show()
```

```python
#Gráfico MERCOSUL
plot_2=dados_media_filtrado.iloc[2, :]
plt.plot(tempo, plot_2, 'green', label='MERCOSUL')
plt.xlabel('Tempo(Anos)')
plt.ylabel('Crescimento do PIB(%)')
plt.grid(True)
plt.legend()
plt.title('Crescimento anual do PIB x Tempo')
plt.show()
```

Assim, conseguimos elaborar gráficos ainda melhores a partir das informações obtidas anteriormente antravés dos gráficos e da analise numérica dos dados, para que assim possamos plotar gráficos conclusivos que reforçem nossa teoria e mostrem visualmente a diferença entre estar em uma organização ou em outra.

```python
#Gráfico OCDE x MERCOSUL
plt.plot(tempo, plot_2, 'green', label='MERCOSUL')
plt.plot(tempo, plot_1, 'blue', label='OCDE')
plt.xlabel('Tempo(Anos)')
plt.ylabel('Crescimento anual do PIB(%)')
plt.grid(True)
plt.legend()
plt.title('OCDE x MERCOSUL')
plt.show()
```

```python
#Gráfico OCDE x Resto do Mundo
plt.plot(tempo, plot_0, 'red', label='Resto do Mundo')
plt.plot(tempo, plot_1, 'blue', label='OCDE')
plt.xlabel('Tempo(Anos)')
plt.ylabel('Crescimento anual do PIB(%)')
plt.grid(True)
plt.legend()
plt.title('OCDE x Resto do Mundo')
plt.show()
```

```python
#Gráfico MERCOSUL x Resto do Mundo
plt.plot(tempo, plot_2, 'green', label='MERCOSUL')
plt.plot(tempo, plot_0, 'red', label='Resto do Mundo')
plt.xlabel('Tempo(Anos)')
plt.ylabel('Crescimento anual do PIB(%)')
plt.grid(True)
plt.legend()
plt.title('Resto do Mundo x MERCOSUL')
plt.show()
```

```python
#Gráfico OCDE x MERCOSUL x Resto do Mundo
plt.figure(figsize=(15,5))
plt.plot(tempo, plot_2, 'g', label='MERCOSUL')
plt.plot(tempo, plot_1, 'b', label='OCDE')
plt.plot(tempo, plot_0, 'r', label='Resto do Mundo')
plt.xlabel('Tempo(Anos)')
plt.ylabel('Crescimento anual do PIB(%)')
plt.grid(True)
plt.legend()
plt.title('OCDE x MERCOSUL x Resto do Mundo')
plt.show()
```

Além disso, o fato de pertencer a OCDE traz algumas vantagens aos seus países membros como a diminuição do custo de captação de emprestimos, alinhamento e ainda facilita acordos bi-laterais entre seus membros, fatores estes que potencializam a economia dos países que obtém essas vantagens, que acabam sendo melhores que as vantagens do MERCOSUL devido ao fato de os países da OCDE serem países de elite no fator econômico, fornecendo assim melhores vantagens que os países ainda em desenvolvimento do MERCOSUL. 

# Conclusão:

Como pode observar-se no gráfico de comparação entre o crescimento do PIB entre a OCDE e o MERCOSUL, é possível afirmar-se que, mesmo que o pico de crescimento do MERCOSUL tenha sido maior que o da OCDE ( fator este que pode ser explicado pelo fato de que a grande maioria dos países da OCDE já são desenvolvidos, enquanto os países do MERCOSUL ainda estão em desenvolvimento, o que faz com que o crescimento do PIB destes países seja menor.) o crescimento do PIB destes países é muito mais constante do que os do MERCOSUL, que por sua vez tem seus valores de crescimento muito instáveis durante os anos. Casos incomuns como a queda gigantesca do crescimento do PIB dos países da OCDE em 2008 devido a uma crise, resultado da falência de um grande banco tradicional americano chamado Lehman Brothers, que afeta muito mais a OCDE devido ao fato de que os Estados Unidos e seus principais parceiros comerciais se encontram dentro desta organização, foram desconsiderados.<br>
Em suma, após a análise dos dados abordados durante este projeto, juntamente com os gráficos obtidos através das informações que foram estudadas, é possível concluir-se que partcipar da OCDE é mais efetivo e gera mais benefícios para o desenvolvimento econômico dos países do que fazer parte do MERCOSUL, sendo assim a organização internacional mais eficaz em gerar crescimento econômico para seus participantes, pois eles passam a ter um crescimento do PIB maior do que se eles estivessem no MERCOSUL.
